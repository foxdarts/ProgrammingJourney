```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report


CCFraud = pd.read_csv("csv_result-creditfruad original.csv") #loads dataset into program.

CCFraud.info() #displays information about dataset.
```

```python
CCFraud.head(10) #shows first 10 lines of dataset so you vcan see values by column.
```

```python
print(CCFraud.columns)  #gives us the column names for hte dataset.

print(CCFraud['class'].unique()) #gives us a list of fraud inputs

print(CCFraud['class'].value_counts()) #tells us how many charges were considered fraud within the dataset
```

```python
CCFraud = CCFraud.drop(['over_draft', 'credit_history', 'purpose', 'Average_Credit_Balance', 'employment', 'location', 'personal_status', 'other_parties', 'property_magnitude', 'other_payment_plans', 'housing', 'job', 'num_dependents', 'own_telephone', 'foreign_worker'], axis = 1) #drops unneeded values from dataset

print(CCFraud.columns) #prints remaining values
```

Prepairing to split and train the data starts here.

```python
labels = LabelEncoder()

CCFraud.iloc[:,6] = labels.fit_transform(CCFraud.iloc[:,6].values) #transform class into values

print(CCFraud['class'].unique()) #prints class values
```

```python
X = CCFraud.iloc[: , 1:8] #splits data into 80%training
Y = CCFraud.iloc[: , 0] #and 20 percent testing

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .6, random_state = 0) #actual split of data into testing and training

SD = StandardScaler() #scaled dataset
X_train = SD.fit_transform(X_train)
X_test = SD.transform(X_test)
```

Decision tree algorithm has given me the best success with this dataset in terms of accuricy.

```python
DETFraud = DecisionTreeClassifier() #decision tree for classification fits dataset best.

DETFraud.fit(X_train, Y_train)

y_pred_decision_tree = DETFraud.predict(X_test)

acc_decision_tree = DETFraud.score(X_train, Y_train) 

print (acc_decision_tree) #decision tree logic.

print (confusion_matrix(Y_test, y_pred_decision_tree)) #shows confusion matrix from trained data to look at

```
